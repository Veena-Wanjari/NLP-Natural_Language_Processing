### Count Vectorizer

Count vectorizer is a part of NLTK(Natural Language Tool Kit library) module in Python. This library helps us to convert text data to a numeric data.

1) We prepare tokens in each vocabulary and then we convert this text data into numeric form.

2) So, we have created a vector in 0 and 1 forms. When there is a tie, it will show 1 or else it will show 0.

3) So, count vectorizer will use the final preprocessing message and then create a vocabulary and depending on the vocabulary it builds the vector.

4) Then we can feed these vectors to our machine learning model.

5) Computer does not understand the text directly so we have to convert into numerical form. So, we make vector space models which helps us to convert text into numerical form.

Before feeding our data to Machine learning model, we need to preprocess the data. Why we need to preprocess??

1) Count vector will convert each text irrespective of their cases(Uppper or lower), plural (code, Codes, codes) and punctuation marks, comma, fullstop will be considered different tokens.

2) The parse matrix is in the form of (a,b) where a is number of rows and b is the particular column name or token name. 

3) If we do not preprocess then this matrix will be a huge parse matrix. This will reduce our computational speed , credibility too, will affect the working or learning of the model, it will reduce the power to extract information from it.
